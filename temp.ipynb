```python

L = 10
num_spins = L * L
spin_config = np.random.rand(num_spins)*2*np.pi
spin_config_z = np.zeros(num_spins)
for i in range(num_spins):
    if np.random.rand() < 0.5:
            spin_config_z[i] = 0
X, Y,Z= np.meshgrid(np.arange(L), np.arange(L),0)
W = np.reshape(spin_config_z, (L, L,1))
Sp = np.reshape(spin_config, (L, L,1))
C = np.cos(Sp)
S = np.sin(Sp)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlim(0, L)
ax.set_ylim(0, L)
ax.set_zlim(-1, 1)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.quiver(X, Y, Z, C, S,W , length=0.7, normalize=True, color="red")

X, Y, Z = np.meshgrid(np.arange(L),int(L/2), 1)

spin_config_x = np.ones(L)
for i in range(L):
    if np.random.rand() < 0.5:
            spin_config_x[i] = -1
U = np.reshape(spin_config_x, (1, L,1))
V = np.zeros((1, L,1))
W = np.zeros((1, L,1))
ax.quiver(X, Y, Z, V, U, W, length=0.7, normalize=True, color="blue")

plt.show()
```
